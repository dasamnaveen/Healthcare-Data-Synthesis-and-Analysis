### Grouping & Aggregation

3. Health Profiles by Gender: What is the average BMI and Cholesterol for each Gender?

```python
gender_vitals = df.groupby('Gender')[['BMI', 'Cholesterol']].mean()

print(gender_vitals)
```

4. Billing Analysis: Calculate the total and average Billing_Amount for each Insurance_Provider

```python
billing_stats = df.groupby('Insurance_Provider')['Billing_Amount'].agg(['sum', 'mean'])

billing_stats.columns = ['Total_Billing', 'Average_Billing']

print(billing_stats)
```

5. Condition Frequency: Which Medical_Condition is most prevalent in each Hospital_Tier?

```python
condition_by_tier = df.groupby('Hospital_Tier')['Medical_Condition'].agg(lambda x: x.value_counts().idxmax())

print(condition_by_tier)
```

6. Blood Type Vitals: Find the median Systolic_BP and Diastolic_BP grouped by Blood_Type.

```python
blood_type_vitals = df.groupby('Blood_Type')[['Systolic_BP', 'Diastolic_BP']].median()

print(blood_type_vitals)
```

7. High-Risk Filter: Create a subset of the data containing only patients aged over 60 with a Systolic_BP greater than 140.

```python
high_risk_patients = df[(df['Age'] > 60) & (df['Systolic_BP'] > 140)]

print(len(high_risk_patients))
```

8. Frequent Exercisers: Filter the dataframe for patients who have "High" Exercise_Freq but are also classified as "Smokers."

```python
active_smokers = df[(df['Exercise_Freq'] == 'High') & (df['Smoker_Status'] == 1)]

print(active_smokers.head())
```

9. Top Billing Patients: Identify the top 5% of patients based on their Billing_Amount.

```python
billing_threshold = df['Billing_Amount'].quantile(0.95)

top_5_percent_billing = df[df['Billing_Amount'] > billing_threshold]

top_5_percent_billing = top_5_percent_billing.sort_values(by='Billing_Amount', ascending=False)

print(f"The minimum amount to be in the top 5% is: {billing_threshold}")
```

10. BP Category: Create a new column BP_Status. If Systolic_BP < 120 and Diastolic_BP < 80, label it "Normal"; otherwise, label it "High".

```python
df['BP_Status'] = np.where((df['Systolic_BP'] < 120) & (df['Diastolic_BP'] < 80), 'Normal', 'High')
```

```python
df['BP_Status']
```

11. BMI Categorization: Create a BMI_Class column (Underweight < 18.5, Normal 18.5–24.9, Overweight 25–29.9, Obese > 30).

```python
bins = [0, 18.5, 24.9, 29.9, 100]
labels = ['Underweight', 'Normal', 'Overweight', 'Obese']

df['BMI_Class'] = pd.cut(df['BMI'], bins=bins, labels=labels)
```

```python
df['BMI_Class']
```

12. Cross-Tabulation: Create a pivot table showing the count of Test_Results (Normal, Abnormal, etc.) across different Medical_Conditions.

14. Outlier Detection: Replace any Glucose_Level values that are more than 3 standard deviations from the mean with the column median.

```python
mean_glu = df['Glucose_Level'].mean()
std_glu = df['Glucose_Level'].std()

median_glu = df['Glucose_Level'].median()
df['Glucose_Level'] = np.where(np.abs(df['Glucose_Level'] - mean_glu) > (3 * std_glu), 
                               median_glu, df['Glucose_Level'])
median_glu
```

## Data Visualization

1. Age Demographics: Create a Histogram of the Age column with 20 bins to see which age groups are most common in the hospital.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Patient Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

2. Financial Skew: Use a Boxplot to visualize the Billing_Amount across different Insurance_Provider types to identify which insurance has the most outliers or the highest median cost.

```python
plt.figure(figsize=(12, 6))

sns.boxplot(
    x='Insurance_Provider', 
    y='Billing_Amount', 
    data=df, 
    hue='Insurance_Provider', 
    palette='Set2', 
    legend=False
)

plt.title('Billing Amount Distribution by Insurance Provider')
plt.xticks(rotation=45)
plt.show()
```

3. Condition Prevalence: Create a Countplot (Bar Chart) showing the number of cases for each Medical_Condition, sorted from highest to lowest.

```python
plt.figure(figsize=(10, 6))
order = df['Medical_Condition'].value_counts().index
sns.countplot(y='Medical_Condition', data=df, order=order, hue='Medical_Condition', palette='viridis', legend= False)
plt.title('Most Common Medical Conditions')
plt.show()
```

4. Gender & Habits: Build a Grouped Bar Chart showing the count of Smoker_Status (Yes/No) separated by Gender.

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Gender', hue='Smoker_Status', data=df)
plt.title('Smoking Status by Gender')
plt.show()
```

5. Test Result Efficiency: Use a Stacked Bar Chart to show the proportion of "Normal," "Abnormal," and "Inconclusive" Test_Results for each Hospital_Tier.

```python
test_counts = df.groupby(['Hospital_Tier', 'Test_Results']).size().unstack()
test_counts.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='Pastel1')
plt.title('Proportion of Test Results by Hospital Tier')
plt.ylabel('Count')
plt.show()
```

6. The BP Scatter: Create a Scatter Plot with Systolic_BP on the x-axis and Diastolic_BP on the y-axis. Color the points by Smoker_Status to see if smokers cluster in the "High BP" zone.

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(x='Systolic_BP', y='Diastolic_BP', hue='Smoker_Status', data=df, alpha=0.6)
plt.title('Systolic vs Diastolic Blood Pressure')
plt.show()
```

7. Vitals Correlation: Generate a Heatmap of the correlation matrix for all numerical columns (Age, BMI, Systolic_BP, Cholesterol, Glucose_Level).

```python
plt.figure(figsize=(12, 8))
# Only include numerical columns
corr = df[['Age', 'BMI', 'Systolic_BP', 'Diastolic_BP', 'Cholesterol', 'Glucose_Level']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Health Vitals')
plt.show()
```

8. Billing vs. Age: Use a Regplot (Scatter plot with a regression line) to see if Billing_Amount increases as patients get older.

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='Age', y='Billing_Amount', data=df, scatter_kws={'alpha':0.1}, line_kws={'color':'red'})
plt.title('Relationship Between Age and Billing Amount')
plt.show()
```

9. Monthly Admissions: Resample the data by month using Visit_Date and create a Line Chart to show the trend of patient admissions over time.

```python
monthly_counts = df.set_index('Visit_Date').resample('ME').size()

plt.figure(figsize=(12, 6))
sns.lineplot(x=monthly_counts.index, y=monthly_counts.values, marker='o', color='royalblue')

plt.title('Monthly Patient Admissions Trend')
plt.xlabel('Visit Month')
plt.ylabel('Number of Admissions')
plt.grid(True, alpha=0.3)
plt.show()
```

10. Weekday Patterns: Create a Bar Chart showing the average Billing_Amount or patient count for each day of the week (Monday through Sunday).

```python
df['Day_of_Week'] = df['Visit_Date'].dt.day_name()

day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

plt.figure(figsize=(10, 6))
sns.countplot(
    x='Day_of_Week', 
    data=df, 
    order=day_order, 
    hue='Day_of_Week',
    palette='pastel', 
    legend=False
)

plt.title('Patient Volume by Day of the Week')
plt.xlabel('Day')
plt.ylabel('Number of Patients')
plt.show()
```
